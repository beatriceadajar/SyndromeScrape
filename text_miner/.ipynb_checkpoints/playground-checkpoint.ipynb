```python
import os, json, nltk, spacy, collections
import numpy as np
import pandas as pd
import matplotlib as plt
from spacy import *
from nltk import *
from nltk.chunk import *
from nltk.stem import *
%matplotlib inline  
```

```python
lemmatizer = WordNetLemmatizer()
df = pd.read_json('semantics.json')
text = df['content'][0]
```

```python
sents = nltk.word_tokenize(text)
words = nltk.pos_tag(sents)
words
```

```python
tree = nltk.ne_chunk(words, binary=True)
iob = nltk.chunk.tree2conlltags(tree)
iob
```

```python
nlp = spacy.load('en')
```

```python
doc = nlp(text)
for ent in doc.ents:
    print(ent.text, ent.start_char, ent.end_char, ent.label_)
```
